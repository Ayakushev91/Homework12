### Задание 1

```python
"""
Напишите функцию, которая классифицирует фильмы из материалов занятия
"""
```

```python
import pandas as pd
```

```python
# Загружаем файлы с фильмами и рейтингами
mov_df = pd.read_csv('/Python/ml-latest-small/movies.csv')
rat_df = pd.read_csv('/Python/ml-latest-small/ratings.csv')
```

```python
# Объединяем датафреймы в один 
ratings_movie = mov_df.merge(rat_df, left_on = 'movieId', right_on = 'movieId', how = 'inner')
ratings_movie.head(20)
```

```python
#группируем фильмы по рейтингу
finally_df = (ratings_movie
    .groupby('title')
    .agg(average_rat = ('rating', 'mean')) 
)
finally_df.head(20)
```

```python
#функция классификации фильмов
def classificator(row):
    if row['average_rat'] <= 2.0:
        row['class'] = 'low'
    elif row['average_rat'] <= 4.0:
        row['class'] = 'mid'
    elif row['average_rat'] >= 4.5:
        row['class'] = 'high'
    return row['class']
```

```python
# Применяем функцию к датафрейму
finally_df['class'] = finally_df.apply(classificator, axis=1)
finally_df.head(50)
```

### Задание 2 

```python
"""
Необходимо написать гео-классификатор, который каждой строке сможет выставить
географическую принадлежность определенному региону. 
"""
```

```python
#Загружаем файл с запросами
keywords_df = pd.read_csv('/Python/keywords.csv')
keywords_df.head(20)
```

```python
# функция гео-классификатор
def geoclass(row):
    for city in row['keyword'].split(' '):
        if city in ('москва','тула','ярославль'):
            row['region'] = 'Центр'
        elif city in ('петербург','псков','мурманск'):
            row['region'] = 'Северо-Запад'
        elif city in ('владивосток','сахалин','хабаровск'):
            row['region'] = 'Дальний Восток'
        else:
            row['region'] = 'undefined'
    return row['region']      
```

```python
# применяем функцию к датафрейму
keywords_df['region'] = keywords_df.apply(geoclass, axis=1)
keywords_df.tail(20)
```

```python
# проверяем результат 
print(keywords_df.loc[keywords_df['region'] != 'undefined'])
```
